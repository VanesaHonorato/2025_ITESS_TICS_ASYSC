## Serie de Fourier

La serie de Fourier permite representar una función periódica \( f(t) \) como una suma infinita de senos y cosenos.

---

### Expresión general

$$
S(t) = a_0
$$

$$
s(f) = a_0 + \sum_{n=1}^{\infty} \left[ a_n \cos(2\pi f_n t) + b_n \sin(2\pi f_n t) \right]
$$

---

### Coeficientes de la serie

Los coeficientes están definidos como:

$$
a_0 = \frac{1}{T} \int_{0}^{T} f(t)\, dt
$$

$$
a_n = \frac{2}{T} \int_{0}^{T} f(t)\cos(2\pi f_n t)\, dt
$$

$$
b_n = \frac{2}{T} \int_{0}^{T} f(t)\sin(2\pi f_n t)\, dt
$$



*Donde \( T \) es el periodo fundamental y \( f_n \) representa las frecuencias armónicas.*


```python
from thinkdsp import CosSignal, SinSignal
pi=3.141516
k=1
n3=3
n5=5
f=0.25

st= CosSignal(freq=f, amp=(2*k)/pi, offset=0)

st2= st+CosSignal(freq=f*n3, amp=-(3*k)/(3*pi), offset=0)

st3= st2+ CosSignal(freq=f*n5, amp=(2*k)/(5*pi), offset=0)
```

```python
from thinkdsp import decorate
wave = st.make_wave (duration=4)
wave.ys+= 0.5

wave2 = st2.make_wave(duration=4)
wave2.ys+=0.5

wave3 = st3.make_wave(duration=4)
wave3.ys+=0.5

st.plot()
st2.plot()
st3.plot()

decorate(xlabel='Time (segundos)', ylabel='Acos(2pift+th)')
import matplotlib.pyplot as plt
plt.grid(True, which ='both')
```
